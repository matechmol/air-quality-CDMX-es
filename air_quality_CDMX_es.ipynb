# **Calidad de aire en CDMX**


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# =========================
# Carga y limpieza de datos
# =========================

# Carga de datos
df = pd.read_csv("air_data_CDMX.csv")

# Primera visualizacion
print(df.head())
print(df.info())
```

```python
# Borrado columnas innecesarias
df = df.drop(columns=["location_name", "location_id", "datetimeLocal", "timezone",
                      "latitude", "longitude", "country_iso", "isMobile",
                      "isMonitor", "owner_name", "provider"])

# Transformando a formato fecha-hora
df["datetimeUtc"] = pd.to_datetime(df["datetimeUtc"])

# Comprobando los valores nulos del dataframe
print(df["parameter"].value_counts())
print(df.isnull().sum())
```

```python
# ===========================
# Visualizacion de los gases
# ===========================

# Agrupando los gases
df_gas = df[df["parameter"].isin(["co", "so2", "nox", "no", "no2", "o3"])]
plt.figure(figsize=(15, 20))
colors = ["skyblue", "lightcoral", "lightgreen", "gold",
          "lightsalmon","lightseagreen"]

# Visualizacion de la concentracion a lo largo del mes
for i in range (len(df_gas["parameter"].unique())):
    plt.subplot(3,2,(i+1))
    df_aux = df_gas[df_gas["parameter"].isin([df_gas["parameter"].unique()[i]])]
    sns.lineplot(data=df_aux, x="datetimeUtc", y="value", color=colors[i])
    plt.title(f"Concentracion del compuesto: {df_gas['parameter'].unique()[i]}")
    plt.xlabel("Fecha")
    plt.xticks(rotation=45, ha="right")
    x_label = plt.gca()
    x_label.tick_params(axis="x", labelsize=10)
    plt.ylabel("Concentracion (ppm)")
    plt.subplots_adjust(hspace=0.4)
    plt.grid()

plt.show()

# Diagramas de caja y bigote para los compuestos
plt.figure(figsize=(15, 20))
for i in range (len(df_gas["parameter"].unique())):
    plt.subplot(3,2,(i+1))
    df_aux = df_gas[df_gas["parameter"].isin([df_gas["parameter"].unique()[i]])]
    sns.boxplot(data=df_aux, x="parameter", y="value", color=colors[i])
    plt.title(f"Diagrama de caja y bigote para: {df_gas['parameter'].unique()[i]}")
    plt.xlabel("Compuesto")
    plt.ylabel("Concentracion (ppm)")

plt.show()

# Comprobando valores estadisticos para los gases
df_gas.groupby("parameter")["value"].describe()
```

```python
# ============================================
# Visualizacion de las particulas suspendidas
# ============================================

df_sp = df[df["parameter"].isin(["pm10", "pm25"])]
plt.figure(figsize=(15, 20))
colors = ["lightcoral", "khaki"]

for i in range (len(df_sp["parameter"].unique())):
    plt.subplot(3,2,(i+1))
    df_aux = df_sp[df_sp["parameter"].isin([df_sp["parameter"].unique()[i]])]
    sns.lineplot(data=df_aux, x="datetimeUtc", y="value", color=colors[i])
    plt.title(f"Particulas suspendidas: {df_sp['parameter'].unique()[i]}")
    plt.xlabel("Fecha")
    plt.xticks(rotation=45, ha="right")
    x_label = plt.gca()
    x_label.tick_params(axis="x", labelsize=10)
    plt.ylabel("Concentracion (g/m)")
    plt.subplots_adjust(hspace=0.5)
    plt.grid()

plt.show()

# Diagramas de caja y bigote para las particulas suspendidas

plt.figure(figsize=(15, 20))

for i in range (len(df_sp["parameter"].unique())):
    plt.subplot(3,2,(i+1))
    df_aux = df_sp[df_sp["parameter"].isin([df_sp["parameter"].unique()[i]])]
    sns.boxplot(data=df_aux, x="parameter", y="value", color=colors[i])
    plt.title(f"Diagrama de caja y bigote para: {df_sp['parameter'].unique()[i]}")
    plt.xlabel("Compuesto")
    plt.ylabel("Concentracion (g/m)")

plt.show()

# Comprobando valores estadisticos para las particulas suspendidas

df_sp.groupby("parameter")["value"].describe()
```

## Conclusiones preliminares
- El compuesto con mayor presencia en el aire es el monoxido de carbono.
- El dataframe no contenia datos nulos, sin embargo si tenia presencia de valores atipicos para los oxidos de nitrogeno (NOx), el dioxido de azufre (SO2) y para ambos casos de particulas suspendidas.
- Las particulas suspendidas de PM 2.5 presentaron una mayor variabilidad que las de PM 10. Esto ocurre dado que las particulas de PM 2.5 son un subgrupo de las de PM 10.
- Este analisis se enfoco en un mes de datos presentes para la zona de La Merced en CDMX.
